```python
from tictactoe import TicTacToe
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 10000
results = np.zeros((n, 3))
for i in range(n) :

    game = TicTacToe()
    game.play(show = False)
    results[i, :] = game.result.reshape(1, 3)
```

```python
evResults = np.zeros((n, 3))
for k in range(n) :
    evResults[k, :] = np.mean(results[:k+1, :], axis = 0)
```

```python
plt.plot(evResults)
plt.show()
```
